```python
df = df.dropna()
```

```python
df.head(3)
```

```python
df.columns
```

```python
#Date of Journey
df['Date_of_Journey'][0]
```

```python
df['Date_of_Journey_day'] = pd.to_datetime(df['Date_of_Journey']).dt.day
df['Date_of_Journey_month'] = pd.to_datetime(df['Date_of_Journey']).dt.month
df['Date_of_Journey_year'] = pd.to_datetime(df['Date_of_Journey']).dt.year
```

```python
df.drop(columns=['Date_of_Journey'], inplace=True)
```

```python
#Dep TIme
print(df['Dep_Time'].min(), df['Dep_Time'].max())
```

```python
df['Dep_Time_hour'] = pd.to_datetime(df['Dep_Time']).dt.hour
df['Dep_Time_minute'] = pd.to_datetime(df['Dep_Time']).dt.minute
```

```python
df.drop(columns=['Dep_Time'], inplace=True)
```

```python
#Arrival TIme
print(df['Arrival_Time'].min(), df['Arrival_Time'].max())
```

```python
df['Arrival_Time_hour'] = pd.to_datetime(df['Arrival_Time']).dt.hour
df['Arrival_Time_minute'] = pd.to_datetime(df['Arrival_Time']).dt.minute
```

```python
df.drop(columns=['Arrival_Time'], inplace=True)
```

```python
df['Duration'].value_counts()
```

```python
#Duration TIme
df['Duration'] = pd.to_timedelta(df['Duration'])
```

```python
print(df['Duration'].min(), df['Duration'].max())
```

```python
df['Duration_hour'] = df['Duration'].dt.components.hours
df['Duration_minute'] = df['Duration'].dt.components.minutes
df['Duration_day'] = df['Duration'].dt.components.days
```

```python
df.drop(columns=['Duration'], inplace=True)
```

```python
#Airline
Airlne = pd.get_dummies(df[['Airline']] , drop_first = True).astype(int)
Airlne
```

```python
#Source
Source = pd.get_dummies(df[['Source']],drop_first=True).astype(int)
Source
```

```python
#Destination
Destination = pd.get_dummies(df[['Destination']],drop_first=True).astype(int)
Destination
```

```python
df= df.drop(['Route','Additional_Info'],axis='columns')
df
```

```python
df['Total_Stops'].value_counts()
```

```python
df['Total_Stops'] = df['Total_Stops'].replace({"1 stop":1,"non-stop":0,"2 stops":2,"3 stops":3, "4 stops":4})
```

```python
df['Total_Stops'].value_counts()
```

```python
df = df.drop(['Airline','Source','Destination'],axis='columns')
df
```

```python
df = pd.concat([df, Airlne, Destination, Source], axis= 1)
```

```python
df.head(3)
```

```python
df.columns
```

```python
target = df['Price']
target
```

```python
df.drop(columns=['Price'], inplace=True)
```

```python
df.head(3)
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train, y_test = train_test_split(df,target, test_size=0.1, shuffle=True)
```

```python
X_train.head(3)
```

```python
y_train.head(3)
```

```python
from sklearn.ensemble import ExtraTreesRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import GridSearchCV,cross_val_score


clf = {
    
    "extrtreeg":{
        "model": ExtraTreesRegressor(),
        "params":{ 

  "n_estimators": [100, 200],
  "max_depth": [10, 20],
  "min_samples_split": [2, 5],
  "min_samples_leaf": [1, 2],
  "max_features": ["sqrt", "log2"],
  "bootstrap": [True]

    }
    },
    
    "RandmForest":{
        "model":RandomForestRegressor(),
        "params":{  
            
  "n_estimators": [100, 200],
  "max_depth": [10, 20],
  "min_samples_split": [2, 5],
  "min_samples_leaf": [1, 2],
  "max_features": ["sqrt", "log2"],
  "bootstrap": [True]
}
    },
    
    "LinReg":{
        "model":LinearRegression(),
        "params":{
  "fit_intercept": [True, False],
  "copy_X": [True],
  "positive": [False, True]
        }
    }
}
```

```python
clf.get("LinReg")
```

```python
scores = []
for model_name, modell in clf.items():
    print("Model Name:", model_name)
    gsc = GridSearchCV(modell['model'],modell['params'],cv=5)
    gsc.fit(X_train,y_train)
    print("Completed Fit")
    scoreb= gsc.score(X_test,y_test)
    cvs = cross_val_score(modell['model'],X_train,y_train)
    scores.append({
        "model_nam":model_name,
        "cross_val_sc": cvs,
        "score":scoreb
    })
```
